```python
%matplotlib inline
### Standard imports
import uproot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqdm
### HEPKIT imports
import hepkit.hepData as hd
import hepkit.hepPlot as hp
import hepkit.hepFormat as hf
### Settings
hf.DefaultSetup()

def PrintKeys(fileName):
    for k in uproot.open(fileName).allkeys():
        print(k)
        
def OpenRoot(fileName,tree):
    rTree = uproot.open(fileName)[tree]
    df = rTree.pandas.df(['*'],flatten=False,namedecode='utf-8')
    return df

# Special trick for string columns
def decodeList(l):
    return [x.decode(encoding='UTF-8') for x in l]
```

```python
import matplotlib.gridspec as gridspec


nbins = 30
xlims = [0,115]

plt.figure(figsize=(8,6),facecolor='white',dpi=100)
gs = gridspec.GridSpec(2, 1,height_ratios=[2,1])
ax1 = plt.subplot(gs[0])
b1,h1,e1 = hp.HepHist(ev1.query('edep_tot_out>0')['edep_tot_out'],nbins=nbins,xlims=xlims,norm=True,yLog=False,xLog=False,color='black',label='Flat')
b2,h2,e2 = hp.HepHist(ev2.query('edep_tot_out>0')['edep_tot_out'],nbins=nbins,xlims=xlims,norm=True,yLog=False,xLog=False,color='C0',label='AFM')
plt.grid(ls='--',color='C7',alpha=0.1)
plt.xlim(xlims)
plt.legend(loc='best')
ax2 = plt.subplot(gs[1])
frac = h2/h1
e0 = np.sqrt(np.power(e1/h1,2.) + np.power(e2/h2,2.))*frac
plt.errorbar(b1-(b1[1]-b1[2])/2.,frac,e0,ls='',marker='o',color='C0')
plt.axhline(1,ls='--',color='black')
plt.grid(ls='--',color='C7',alpha=0.1)
plt.xlim(xlims)
plt.ylim(0.7,1.3)
plt.xlabel(r'Energy deposit outside PTFE [keV]')
plt.sca(ax1)
plt.show()
```

```python
nbins = 30
xlims = [1e-1,1e4]
# xlims = [0.1,115]
yLog = False
xLog = True

plt.figure(figsize=(6,6),facecolor='white',dpi=100)
hp.HepHist(ev1.query('edep_tot_out>0')['edep_lead_out'],nbins=nbins,xlims=xlims,norm=False,yLog=yLog,xLog=xLog,color='C3',label=r'$^{206}$Pb')
hp.HepHist(ev1.query('edep_tot_out>0')['edep_alpha_out'],nbins=nbins,xlims=xlims,norm=False,yLog=yLog,xLog=xLog,color='C0',label=r'$\alpha$')
hp.HepHist(ev1.query('edep_tot_out>0')['edep_other_out'],nbins=nbins,xlims=xlims,norm=False,yLog=yLog,xLog=xLog,color='C8',label='Sput. ions')
hp.HepHist(ev1.query('edep_tot_out>0')['edep_tot_out'],nbins=nbins,xlims=xlims,norm=False,yLog=yLog,xLog=xLog,color='C7',label='Total')
plt.grid(ls='--',color='C7',alpha=0.1)
# plt.ylim(1,1e5)
plt.xlabel(r'Energy deposit outside PTFE [keV]')
plt.legend(bbox_to_anchor=(0, -0.15, 1, 0), loc=2, ncol=2, mode="expand", fontsize=14, borderaxespad=0, frameon=False)
plt.title("100 nm implantation | Flat surf.")
plt.show()
```
